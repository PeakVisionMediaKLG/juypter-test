Matplotlib ist ein Modul zur Visualisierung von Daten. Es ist nahtlos in Python einzubinden und enthält eine unglaubliche Vielfalt von Methoden, es ist kostenlos und quelloffen. In diesem Tutorial ist es natürlich nicht möglich, alle Facetten von Matplotlib zu demonstrieren. Wir werden uns auf wichtige Anwendungen beschränken und mittels Beispielen lernen. Unter https://matplotlib.org/2.0.2/gallery.html findet man eine große Zahl von Beispielen, die zeigen, was alles machbar ist. Oft ist es am sinnvollsten, statt alle Parameter der zahlreichen Methoden vorab zu memorieren, sich Beispiele anzuschauen und für eigene Zwecke zu modifizieren. Eine besondere Bedeutung hat Matplotlib im ML. Hier ist es oft von entscheidender Bedeutung, sich zur Auswahl des richtigen Arbeitsalgorhithmus die Rohdaten oder auch transformierte Daten zu visualisieren und ebenso auch Ergebnisse graphisch darzustellen. Meist wird das Untermodul pyplot verwendet. Ähnlich wie NumPy ist auch Matplotlib nicht Bestandteil der Kernpythons. In ML Plattformen wie Anaconda ist es integriert, ansonsten muß man es extra installieren. Wir werden jetzt zunächst mit Linien-Plots arbeiten und zunächst einen simplen Plot erzeugen. 

````python
import matplotlib.pyplot as plt   #üblicher Alias, keiner möchte den vollen Namen ausschreiben
import numpy as np  #fakultativ, matplotlib arbeitet auch mit Python-Listen
werte=np.array([2,5,4,6,3])
plt.plot(werte) #dient zum Festlegen der Art des Diagramms, plot macht ein einfaches Punkt- oder Liniendiagramm
# werte dient in diesem Fall als Array der y-Werte, die x-Werte werden automatisch erzeugt !
plt.show() #Befehl zur Darstellung des Plots
````

Wie wir sehen, gibt es kaum eine einfachere und schnellere Möglichkeit, einfache Diagramme darzustellen. Die Art der Darstellung lässt sich über einen Formatparameter bestimmen.

````python
werte=np.array([2,5,4,6,3])
plt.plot(werte,"r") #  r steht für die Linienfarbe rot
plt.show()
plt.plot(werte,"og") # o steht für Kreise als Punktmarker, g für die Farbe grün dieser Kreise
plt.show()
````

Lässt man den ersten "plt.show" Befehl weg, so werden die Linien übereinandergelegt.

````python
werte=np.array([2,5,4,6,3])
werte1=np.array([3,4,5,6,5])
plt.plot(werte,"r") #  r steht für die Linienfarbe rot
plt.plot(werte,"og") # o steht für Kreise als Punktmarker, g für die Farbe grün
plt.plot(werte1,"b") #  b steht für die Linienfarbe blau
plt.plot(werte1,"xr") # x steht für X als Punktmarker, r für die Farbe rot
plt.show()
````

Auch für diese zusätzlichen Einträge ist eine umfangreiche Formatierung möglich, siehe Dokumentation. Wichtig ist die Möglichkeit der Begrenzung der Achsen auf Minimum- und Maximumwerte.

Achsen lassen sich verschieben, die Markierungen auf den Achsen formatieren. Hierzu bietet sich die gca_Funktion an, die auf die Achsen verweist. <br><b>Die Syntax ist ax=plt.gca()</b> Damit können wir jetzt auf das Achsenobjekt und seine Eigenschaften zugreifen.


Das Achsen-Objekt kennt 4 Achsen, sog. "spines" , die mit "top","bottom","right" und "left" bezeichnet werden. Diese haben Methoden und Eigenschaften und können damit einzeln verändert werden.

````python

x=np.linspace(-2*np.pi,2*np.pi,70,endpoint=True)
F1=np.sin(2*x) #macht diese Werte für alle x , siehe NumPy
F2=np.cos(x)
axis=plt.gca()
axis.spines["top"].set_color("red") #Formatieren der Einzelachse
axis.spines["bottom"].set_color("blue")
axis.spines["right"].set_color("green")
axis.spines["left"].set_color("black")
plt.plot(x,F1) #hier Plot von x-Werten aus x und y-Werten aus F1
plt.plot(x,F2)
plt.show()
````

````python
x=np.linspace(-2*np.pi,2*np.pi,70,endpoint=True)
F1=np.sin(2*x)
F2=np.cos(x)
ax=plt.gca()
ax.spines["top"].set_color("none")
ax.spines["bottom"].set_color("blue")
ax.spines["right"].set_color("none")
ax.spines["left"].set_color("red")
ax.xaxis.set_ticks_position("bottom") #auf welche Achse sollen die Markierungen (ticks) für x kommen
ax.yaxis.set_ticks_position("left") #auf welche Achse sollen die Markierungen für y kommen
ax.spines["left"].set_position(("data",0)) #setze die Y Achse auf den Wert 0 , nehme Koordinaten wie in Daten
plt.plot(x,F1)
plt.plot(x,F2)
plt.show()

plt.xticks(np.arange(10))
plt.plot(x,F2)
plt.xticks([-np.pi,np.pi]) #ticks selbst setzen auf feste Werte
plt.show()

x=["Hund","Katze","Bär"] #ticks setzen auf Kategorien
y=[30,20,120]
plt.plot(x,y,"o")
plt.show()
````

Sowohl für die Achsen (spines) als auch für die Achsenmarken (ticks) gibt es zahlreiche Formatierungsmöglichkeiten.

Sekundäre Achsen sind auch möglich. Man benutzt hier die "subplot" Methode und definiert so verschiedene Abbildungen (figures) und Achsen. Wenn "subplots" ohne Parameter verwendet wird, wird wie hier nur eine Abbildung (in dem Fall nicht mehrere benötigt) und eine Achse zurückgegeben. "twinx" macht ein neues Achsenelement in derselben Abbildung, die x Achse wird gemeinsam genutzt. Mit "twiny" wird hingegen die y-Achse gemeinsam genutzt.


````python
werte=np.array([2,5,4,6,3])
werte1=np.array([33,40,5,60,50])

fig,ax1=plt.subplots()
ax1.plot(werte,"r") #  r steht für die Linienfarbe rot
ax1.plot(werte,"og") # o steht für Kreise als Punktmarker, g für die Farbe grün
ax2=ax1.twinx()
ax2.plot(werte1,"b") #  b steht für die Linienfarbe blau
ax2.plot(werte1,"xr") # x steht für X als Punktmarker, r für die Farbe rot
plt.show()
````

````python
xwerte=np.array([2,5,4,6,3])
xwerte1=np.array([33,40,5,60,50])
fig,ax1=plt.subplots()
ax1.plot(xwerte,"r") 
ax1.plot(xwerte,"og") 
ax2=ax1.twiny()
ax2.plot(xwerte1,"b") 
ax2.plot(xwerte1,"xr")
## ax=plt.gca() #obere Achse ohne Ticks##
## ax.xaxis.set_ticks_position("bottom")##
plt.show()
````

Indem man verschiedenen Plots "label" zuweist und "legend" verwendet, kann man eine Legende einfügen an verschiedenen Positionen.
"annotate" erlaubt das Einfügen von Kommentaren. Dabei bezeichnet "xy" die Lage der Pfeilspitze, "xytext" die Lage des Textes, "fontsize" erklärt sich von selber, "arrowprops" ist ein Dictionary der Pfeileigenschaften, welches ebenfalls wieder Formatierungsmöglichkeiten für den Pfeil enthält.("shrink" steht z.B. für die Verkürzung des Pfeils in Relation zur Länge des größtmöglichen Pfeils, der den Text und die Pfeilspitze verbindet , hier 20%).

````python
werte=np.array([2,5,4,6,3])
werte1=np.array([33,40,5,60,50])

plt.plot(werte,"r",label="Werte") #  r steht für die Linienfarbe rot
plt.plot(werte,"og") # o steht für Kreise als Punktmarker, g für die Farbe grün

plt.plot(werte1,"b",label= "Werte1") #  b steht für die Linienfarbe blau
plt.plot(werte1,"xr") # x steht für X als Punktmarker, r für die Farbe rot
plt.legend(loc="upper left")
plt.annotate("Wichtiger Punkt",xy=(3,60),xytext=(1,50),fontsize=12,arrowprops=dict(facecolor="red",shrink=0.2)) ##shrink = 0##
plt.show()
````

"subplot" erlaubt das Abbilden mehrerer Einzelplots. Man kann mit 2 Parametern die Anzahl von Reihen, Spalten angeben, und ein dritter Parameter beschreibt dann, wo der jeweilige Plot, in deisem Feld abgebildet werden soll. Ein Beispiel zeigt dies. Für 3 Plots zeigen wir einfach einen Text an. Für einen Plot ein Tortendiagramm (davon später mehr).

````python
werte=np.array([2,5,4,6,3])
werte1=np.array([33,40,5,60,50])
fig, ax = plt.subplots(3, 2, figsize=(10,10)) ##figsize bestimmt eine identische Größe für alle Einzelplots
fig.tight_layout() ### genug Abstand um alle Achsen mit Beschriftung abzubilden ###

plt.subplot(3,2,1) #leer
plt.text(.2,.5,"subplot(1)") #Text startet bei x,y

plt.subplot(3,2,2) #3 Reihen, 2 Spalten, Abbildung 2
plt.plot(werte,"r",label="Werte") 
plt.plot(werte,"og")

plt.subplot(3,2,3)
plt.text(.2,.5,"subplot(3)")#leer

plt.subplot(3,2,4) #leer
plt.text(.2,.5,"subplot(4)")

plt.subplot(3,2,5) #3 Reihen, 2 Spalten, Abbildung 5
plt.plot(werte1,"b",label= "Werte1") 
plt.plot(werte1,"xr")

plt.subplot(3,2,6) #leer
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.pie(students, labels = langs,autopct='%3.1f%%')

plt.show()
````

Hier wird ein Feld von Abbildungen mit 3 Reihen und 2 Spalten aufgebaut, die Plots 2,5 und 6 sind dann mit Einzelplots belegt, die anderen nur durch Texte, wodurch man sieht, wie die einzelnen Plots durchnumeriert sind. <br><br> Auch logarithmische Achsenskalierung ist möglich, wie das untere Beispiel zeigt.

````python
x=np.arange(0,100)
fig=plt.figure(figsize=(10,10))

ax=fig.add_subplot(2,1,1) #wir fügen in das fig Objekt einen Subplot ein und bearbeiten dessen ax Objekt
ax.set_yscale("log") #y log
plt.plot(x,x**2,x,x**3)
ax1=fig.add_subplot(2,1,2) #wir fügen in das fig Objekt einen Subplot ein und bearbeiten dessen ax Objekt
ax1.set_yscale("log") #doppelt log
ax1.set_xscale("log")
plt.plot(x,x**2,x,x**3)
plt.show()
````

Bevor wir zu anderen Plots wie Streudiagrammen,Balkendiagrammen, Säulendiagrammen... kommen, möchten wir noch kurz auf die meshgrid-Methode eingehen, da diese in ML - Anwendungen nicht selten Verwendung findet.
```meshgrid``` erzeugt aus zwei Arrays ein Gitter von Punkten.

````python
x_values=np.arange(9)
y_values=np.arange(12)
x,y=np.meshgrid(x_values,y_values)
plt.plot(x,y,linestyle="none",c="red",marker=".")
plt.show()
````

Interessant wird die Methode, wenn man als Z-Wert eine Funktion verwendet, die dann die Darstellung der meshgrid-Punkte beeinflusst. Es entsteht dann ein Kontourplot.

````python
import random
fig=plt.figure(figsize=(10,10))
x_values=np.linspace(2,10,200)
y_values=np.linspace(2,10,200)
x1,y1=np.meshgrid(x_values,y_values)
Z=np.log(x1)  + np.sin(y1)

plt.imshow(Z,origin="lower") #imshow belegt die Rasterpunkte mit Z
plt.title('Kontourplot log(x) + sin(y)')
plt.xlabel('x ')
plt.ylabel('y ',rotation="horizontal")
ax=plt.gca()
ax.grid(color="none")
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,10))
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sin(2*X) + np.cos(2*Y) #min 2 max 2
cp = plt.contour(X, Y, Z) # eine andere Möglichkeit eines Kontourplots
plt.colorbar(cp) #zeigt Schlüssel für Z an in Leiste an
plt.title('Kontourplot sin(2x) + cos(2y)')
plt.xlabel('x ')
plt.ylabel('y ',rotation="horizontal")
plt.show()
````

Eines der wichtigsten Plots im ML stellt das Streudiagramm (Scatterplot) dar, weil beim ML oft Punktwolken zur Datenvisualisierung dargestellt werden. Hier ein Beispiel mit Clustern, wir machen mit dem ML Modul make_blobs von sklearn 3 Punktwolken (davon später mehr), weisen sie 3 Clustern zu und plotten sie in ein Streudiagramm mit Farbzuordnung zu den Clustern: 

````python
import matplotlib.pyplot as plt
import numpy
 
# Mache 2 dimensionale Wolken mit 3 Clustern
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=2)

# Benutze colormap
colormap = numpy.array(['r', 'g', 'b'])


plt.title("3 Wolken")
plt.scatter(X[:,0], X[:,1], s=50, c=colormap[y], marker="p") #die Farben richten sich nach dem y Wert
plt.show()
````

Mit einem Histogrammbeispiel möchte ich die Kurzbeschreibung des Matplotlib Moduls abschliessen. Wir erzeugen hier ein Histogramm, indem wir normalverteilte Zufallszahlen produzieren mit ```loc``` als Mittelwert und ```scale``` als Standardabweichung. Die Anzahl der Zufallszahlen ist ```size```. Das ```random``` Modul von Numpy ist in vielen Belangen sehr ähnlich zum Kernpython random-Modul. Näheres beschreibt die Numpy Dokumentation. <br>
Diese Zahlen werden dann in ```bins``` gleich grosse Bereiche sortiert und die Häufigkeit von Zahlen, die in den jeweiligen Bereich fallen, als Säule angezeigt.

````python
fig=plt.figure(figsize=(8,6))
numbers=np.random.normal(size=100000,loc=-1,scale=2) ## "size" Zufallszahlen normalverteilt Mittelwert -1 , Standardabweichung 2
print(numbers[:10]) #erste 10 Zufallszahlen
plt.hist(numbers,bins=50) ##diese werden in Histogramm geordnet mit "bins" gleich grossen Bereichen 
plt.show()
````

Hier noch ein Beispiel mit Formatierung einer Säule und einer Kurve für die erwartete Verteilung:

````python
fig=plt.figure(figsize=(12,7))
np.random.seed(1)
x=np.random.normal(size=10000,loc=100,scale=20)
loc,scale=100,20   
num_bins = 50


n, bins, patches = plt.hist(x, num_bins, color ='green')
# n ist Array der Häufigkeitswerte aller bins
# bins ist Array der linken Begrenzung aller bins
# patches ist Array aller gezeichneten Balken
print(plt.getp(patches)) #alle Eigenschaften der patches
plt.setp(patches[30],"fc",(0.8,0.2,0.7,1.0),"ec",(0,0,0,1),"lw",2) # der  patch 30  wird umgefärbt
                    #fc facecolor , ec edgecolor, lw linewidth



# Macht Linie der erwarteten Verteilung
y = ((1 / (np.sqrt(2 * np.pi) * scale)) *
     np.exp(-0.5 * (1 / scale * (bins - loc))**2))

max_häufigkeit=np.max(n)
ymax=np.max(y)
y = y/ymax*max_häufigkeit #normierung auf Maximum der Verteilung

plt.plot(bins, y, linestyle='-', linewidth=2, color ='red') #Kurve der erwarteten Verteilung
  

  
plt.show()
````
