## Beurteilung der Performance eines Lernprogramms
<img width=400 height=400 class="imgright" src="Images/Error.jpg">


In diesem Kapitel sehen wir uns die Beurteilung der Performance, z.B. der Fehlerquote, von Lernalgorithmen an. Die Ergebnisse eines ML Programms sind in der Realität selten 100% korrekt. Es ist zur Beurteilung der Performance eines solchen Programmes daher notwendig, Kriterien für die Korrektheit des Ergebnisses zu erstellen, um verschiedene Ansätze vergleichen zu können. Grundlage dazu ist die sogenannte Fehlermatrix oder auch Konfusionsmatrix oder Kontingenztabelle, die wir im Laufe des Kapitels ansehen werden. <br><br>Was ist ein Klassifikator? Ein ML Algorithmus, der versucht Objekte einer bestimmten Kategorie von mehreren möglichen zuzuordnen.<br><br>Nehmen wir an, unser Klassifikator müsste die hier dargestellten Beispielobjekte (blaue Punkte (positiv) und braune Punkte (negativ) - die tatsächlichen Eigenschaften des Objekts werden farblich gekennzeichnet) anhand ihrer Position im Koordinatensystem einer blauen Klasse (positive Klasse) und einer braunen Klasse (negative Klasse) zuordnen (vorausgesagte Zugehörigkeiten). Dafür hat der Klassifikator im Beispiel eine Grenzlinie im Koordinatensystem erstellt und damit einen blauen Bereich und einen braunen Bereich abgetrennt. Die Punkte in den jeweiligen Bereichen ordnet er dann der Kategorie blau oder braun zu.


<img width=800 height=800 class="imgright" src="Images/Kategorisieren.png">

Wir können sehen, daß der Klassifikator nicht fehlerfrei gearbeitet hat. Von den 20 blauen Punkten sind 3 im braunen Bereich falsch zugeordnet, von den 16 braunen Punkten 5 im blauen Bereich. Wir können dies jetzt in eine Fehlermatrix übertragen. <b>Dabei ist die Kategorie blau als positiv bewertet.</b><br>
Da die Fehlermatrix sich auf blaue Punkte bezieht, werden blaue Punkte generell als positiv (wahr, gehören der echten Kategorie an) bezeichnet, braune Punkte als negativ.<br>

Betrachten wir zum Beispiel einen Test, der bei einem Patienten voraussagen soll, ob er an einer gefährlichen übertragbaren Krankheit leidet oder nicht. Was müßte bei diesem Test besonders hoch bewertet werden, wenn wir das Ergebnis <b>"der Test ist positiv"</b> betrachten.<br> Wenn das Ergebnis <b>falsch positiv</b> ist, wird bei einem Patienten, der nicht erkrankt ist, die Krankheit diagnostiziert. Was könnte das für den Patienten und was für die Gesellschaft bedeuten. Der Patient hat dann vielleicht mit unangenehmen oder gar gefährlichen Maßnahmen zu rechnen, z.B. Quarantäne oder der unnötigen Behandlung mit nebenwirkungsreichen Medikamenten, die Gesellschaft wäre vor einer Infektion gewarnt und könnte entsprechend reagieren, auch wenn das objektiv nicht nötig wäre.<br>
Wenn das Ergebnis hingegen <b>falsch negativ</b> wäre, würde der Patient nicht behandelt, obwohl dies nötig wäre, und die Gesellschaft müßte mit der Ausbreitung der Infektion rechnen. Was schlimmer wäre, hängt ganz von der genauen Situation ab, der Gefährlichkeit der Erkrankung, der Nebenwirkungsrate der Behandlung...

Wichtig ist die im Alltag nicht immer klare Unterscheidung von Präzision (precision) und Genauigkeit (accuracy). Für die Bedeutung im Berich ML demonstrieren wir dies hier an einem Beispiel mit Zielscheiben:<br><img width=800 height=800 class="imgright" src="Images/Präzision_Genauigkeit.png">



Daneben gilt es beim Maschinellen Lernen bei allen Verfahren zu beachten, daß man das Modell in der Lernphase jeweils zu wenig oder zu stark an die Daten anpassen kann, was dann entweder die Daten unzureichend nutzt oder bei der Voraussage von unbekannten Werten zu Problemem führen kann, oder zumindest die Berechnung unnötig verkomplizieren kann. Man nennt Ersteres <b> Underfitting</b>, Letzteres <b> Overfitting</b> und das Optimum bezeichnet man als <b> Sweet Spot</b>.
Dies Optimum muss für alle Algorithmen jeweils bestimmt werden. Schematisch zeigen die folgenden Abbildungen den Sachverhalt.
