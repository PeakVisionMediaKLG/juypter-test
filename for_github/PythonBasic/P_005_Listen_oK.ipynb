Die Liste ist in Python der Universalcontainer, es ist eine geordnete Gruppe von Elementen beliebigen Typs. Man kann alles als Element in Listen eintragen. In anderen Sprachen sind solche kombinierte Datentypen als Arrays oder Felder bekannt. Meist können sie nur Werte eines bestimmten Typs aufnehmen. Python-Listen sind aber weitaus flexibler.<br><b>Die Haupteigenschaften von Python-Listen:
- Sie sind geordnet. (sequentieller Datentyp)
- Sie enthalten beliebige Objekte. (zusammengesetzter Datetyp, nicht fundamental, da die Teile keine Listen sein müssen)
- Auf Elemente einer Liste kann über einen Index zugegriffen werden.
- Sie können beliebig verschachtelt werden. So können Sie andere Listen als Unterlisten enthalten.
- Ihre Größe ist variabel, Listen können wachsen oder schrumpfen zur Laufzeit
- Sie sind veränderlich. Die Elemente einer Liste können geändert werden.

Listenobjekte werden in eckige Klammern eingeschlossen und die Elemente durch Kommata getrennt. </b>Einige Beispiele für Listen:



````python
l = [] # eine leere Liste
l = [3,4,-3] # eine Liste von Integern
l = [2.3,-4,5 + 6j] # eine Liste mit verschiedenen numerischen Typen
l = ["Hallo",4,2.1,False] # von allem etwas
l = [1,2,["hallo","bla"],(2,3)] # eine verschachtelte Liste mit einer Unterliste und einem Tupel, wie ist die Länge?
print(f' [1,2,["hallo","bla"],(2,3)] hat Länge {len(l)}') #nur die oberste Ebene der Elemente bestimmt die Länge
l = [1,[2,3,[5,"Möchte ich haben"],[5,7]]] # eine mehrfach verschachtelte Liste Länge?
print(f' [1,[2,3,[5,"Möchte ich haben"],[5,7]]] hat Länge {len(l)}')
````

Man kann Listen ausschneiden, wie schon gezeigt. Der Ausschnitt kann auch wie ein einzelnes Listenelement überschrieben werden. Man kann aber einen Ausschnitt nur wieder durch eine <b>Liste</b> ersetzen, ein Element als Ersatz muß daher so geschrieben werden ```[Element]```. 

````python
l = [0,1,2,3]
teil = [l[1:3]] # nur das Element mit Index 1 und 2
print(f" Der Ausschnitt ist die Liste: {teil} vom Typ: {type(teil)}") #der Ausschnitt ist vom Typ Liste siehe unten
l[2:] = [5] #Achtung,auch wenn man diesen Ausschnitt nur durch ein Listenelement ersetzt, muß man daraus eine Liste machen
print(l)
````

Nun zu den speziellen Listen Funktionen:
<b>
- list.append(Element zum Anhängen)<br>
Ein neues Listenelement an die Liste anhängen. Wie ```+```.
<br><b>Wichig ist, daß die Funktion die Originalliste direkt bearbeitet und verändert. Sie gibt kein Ergebnis (bzw. ```None``` als Nicht-Wert) zurück. Dieses direkte Bearbeiten der Originalliste nennt man "inplace" Funktion.</b><br></b>
    ```+``` gibt allerdings die verlängerte Liste zurück. Da ```+``` deutlich langsamer ist als append() sollte man Letzteres verwenden.


````python
l = [1,2,3,4]
l.append(7) #7 ist Element!!! könnte aber alles sein!! auch Liste s.u.
print(l) #Originalliste verändert
l = l + [8] #Einzelelement auch hier wieder als Liste eingeben s.o.
print(l)
ergebnis = l.append([10]) #Rückgabewert von append()
print(f"Rückgabewert von l.append(10): {ergebnis} , die Originalliste l ist jetzt: {l}")
````

Will man stattdessen eine Liste anhängen und nicht nur ein Element verwendet man <b>
- list.extend(Liste zum Anhängen) Dies verhält sich wie +=,<br></b> allerdings ist extend() inplace, dagegen gibt +=  die verlängerte Liste zurück.+= ist massiv langsamer, man sollte immer extend() verwenden. Man muß einen <b>kombinierten Datentyp </b>(der iterierbar ist, den man also durchfahren kann) anhängen, ein einfacher Datentyp macht einen Fehler.
    

````python
l = [1,2,3]
l2 = [4,5]
l.extend(l2) #l2 ist Liste!!! und muß Liste sein
print(l)
l += [8] #Einzelelement auch hier wieder als Liste eingeben s.o.
print(l)
ergebnis = l.extend([10]) #Rückgabewert von append()
print(f"Rückgabewert von l.extend([10]): {ergebnis} , die Originalliste l ist jetzt: {l}")
#l.extend(5) #macht Fehler: TypeError: 'int' object is not iterable
````

Man kann auch Elemente in die Liste an bestimmten Stellen und nicht nur am Ende einfügen. Das geht mit <br><b>
- list.insert(Index, Element)<br></b>
Am Listenindex "Index" wird das Element "Element" eingefügt. Es wird immer vor dem angegebenen Index eingefügt. l.insert(0,Element) fügt also am Anfang ein. Insert ist ebenfalls inplace. Das Element kann alles sein.

````python
l = [1,2,3,4]
l.insert(2,["Hallo",3]) #vor Index 2
print(l)
````

Das Gegenteil geht mit <br><b>
- list.remove(Wert)<br></b>
Das erste Element, dessen Wert "Wert" ist, wird aus der Liste inplace entfernt. <b>Gibt es ein solches Element nicht, entsteht ein Fehler.</b> Testen kann man vorher natürlich mit dem ```in```-Operator so: ```Wert in Liste```. Dieser Test gibt True oder False zurück, je nachdem, ob der Wert in der Liste vorkommt oder nicht.

````python
l = [2,3,2,4,3]
test = 3 in l
print(f"Ist 3 in l? {test}")
l.remove(3)
print(l)
````

Verwand mit remove ist <br><b>
- list.pop([Index])<br></b>
Das Element am Index "Index" wird entfernt und zurückgegeben! Also einmal inplace Veränderung der Liste aber auch Rückgabe des ausgeschnittenen Elements. Fehlt die Angabe des Index wie hier "l.pop()" wird das letzte Element bearbeitet.

````python
l = [1,2,3,4,5]
ausgeschnittenes_element = l.pop(3)
print(f" Ausgeschnitten: {ausgeschnittenes_element} Liste jetzt: {l}")
ausgeschnittenes_element = l.pop()
print(f" Ausgeschnitten letztes Element: {ausgeschnittenes_element} Liste jetzt: {l}")
````

Löschen eines Elements oder eines Ausschnitts der Liste an einem bestimmten Index oder einem Bereich ```[Start,Stop,Schrittweite]``` geht mit<br><b>
- del liste[index] oder
- del liste[start,stop,schrittweite]</b>

````python
l = [1,2,3,4,5,6,7,8]
del l[1:3:2]
print(l)
del l[2]
print(l)
````

<b>- list.count(Element) <br></b>
gibt zurück, wie oft "Element" in der Liste vorkommt.


````python
l = [2,3,5,3,6,3]
wieviel_dreier = l.count(3)
print(f"Wieveile Dreier? {wieviel_dreier}")
wieviel_siebener = l.count(7)
print(f"Wieveile Siebener? {wieviel_siebener}")
````

<b>- list.index(Wert[, Start[, End]])<br></b>
Gibt den Index des ersten Elements in der Ausgangsliste zurück, dessen Wert "Wert" ist. Mit Start und Ende kann man den Suchbereich bestimmen, wie beim Ausschneiden. Bei einem Suchbereich ist Start dann obligatorisch, End hat als Standardwert das Ende der Liste, falls dieser Wert fehlt.


````python
l = [2,3,4,3,5,6,3]
erstes_mal= l.index(3)
print(f"In der Liste {l} kommt 3 zum ersten mal bei Index {erstes_mal} vor")
such_nur_teil = l.index(3,4,7) #Achtung start und ende kommasepariert nicht [4:7]!
print(f"Im Ausschnitt [4:7] kommt 3 zum ersten mal bei {such_nur_teil} vor")
````

Mit key kann man angeben, wonach sortiert werden soll, was wir bis jetzt nur unzureichend verstehen können (später mehr dazu), reverse bestimmt, ob auf- oder absteigend sortiert werden soll.<br>

Das zweitere macht man mit<br>
<b>- neue_liste = sorted(list,key=None,reverse=False)<br></b>
Gibt man keinen key an, wird nach der in Python eingebauten Reihenfolge sortiert. Gibt man nicht an ```reverse=True``` wird aufsteigend sortiert. Am Schluss ein Beispiel für den Einsatz von key.
````python
l = [5,3,7,2,9,4]
l.sort()
print(f"Sortiert ohne reverse=True: {l}")
l.sort(reverse=True)
print(f"Sortiert mit reverse=True: {l}")
l1 = ["6","3","9","Vier","vier"]
nl = sorted (l1)
print(f'["6","3","9","Vier","vier"] mit Python interner Reihenfolge sortiert\n (Zahlen vor Buchstaben, Groß vor klein) {nl}')
print(f"mit sorted() ist die Ausgangsliste unverändert! {l1}")

def zwei(x):
    return x[1]
l = ["zwei","Bla","foo","vier"]
print(f"{l} wurde nach dem zweiten Buchstaben jedes Eintrags sortiert: {sorted(l,key=zwei)}") 
````

Es fehlt jetzt noch ein etwas komplizierteres Kapitel. Wie kopiert man Listen? Hat die Liste nur eine Ebene und ist nicht verschachtelt, ist es einfach mit <br>
<b>- list.copy()<br></b>
Dies gibt eine Kopie der Liste zurück.

````python
l1 = [1,2,3]
l2 = l1.copy()
print(l2)
print(id(l1),id(l2))
l2[1]="neuer Wert"
print(l1,l2)
########## Aber Achtung!!!!!!!!!!
l3 = [1,2,3]
l4 = l3
l4 [1]="anders"
l3 [0]="ebenfalls anders"
print(l3,l4) ##########!!!!!!!!!!
````

Die Benutzung der copy() Funktion erstellt eine Kopie der Liste l1 und legt sie in l2 ab. (Ginge auch mit l2 = l1 [ : ]). l2 ist von l1 völlig unabhängig, wie man mit der id() Funktion zeigen kann. Daher wird l1 nicht verändert, wenn man l2 intern umstrukturiert. (Zeile 5)
Setzt man zwei Listen aber nur gleich, wie in Zeile 9, wird eine neue <b>zweite Referenz</b> auf die Liste erzeugt, keine neue Liste. <b>Ändert man die Liste über eine der Referenzen, wird damit die einzige Liste, die vorhanden ist, geändert, und beide Referenzen zeigen nun auf diese geänderte Liste.</b> Dies ist der Preis des referentiellen Objektspeichermodells von Python und kann am Anfang zu Verwirrung führen.

Noch komplizierter wird es bei verschachtelten Listen. <br><br><img width=900 src="Images/VerschachtelteListen.png" />

Der Bezeichner L verweist auf einen Block mit zwei weiteren internen Referenzen, wovon eine auf den Wert 1 zeigt. Die zweite Referenz dieses Blocks verweist auf einen inneren Block, wo die einzelnen Elemente der Unterliste referenziert werden.

Wenn wir jetzt ```L1 =  L.copy()``` verwenden, wird der nur erste Referenzblock mit seinen Referenzen kopiert, <b>nicht aber der interne zweite Block mit seinen Referenzen.<br><br><img width =1100 src="Images/VerschachtelteListen2.png" />

Machen wir ```L1[0] = 5``` wird die Referenz auf 1 von der auf 5 getrennt, wie bei einer einfachen doppelt referenzierten Variablen (s.Kapitel Variablen):<br><br><img width=1100 src="Images/VerschachtelteListen3.png" /> 

Was passiert, wenn wir L1[1][1] auf 6 setzen? <br><br><img width=1100 src="Images/VerschachtelteListen4.png" /> 

Damit ist klar, daß sich das neue Element sowohl für L als auch für L1 geändert hat. Wollen wir dies ändern, müssen wir das Modul copy benutzen und eine deepcopy machen. Dann haben wir die folgende Situation: Alle Refrenzen bis zur tiefsten Schicht werden kopiert. Jetzt ändert sich nur L1.<br><br><img width=1100 src="Images/VerschachtelteListen5.png" /> <br> Statt zu schreiben ```L1 = copy(L)``` wäre die folgende Befehlsfolge nötig. 

````python
import copy #nur einmal im Programm nötig, normalerweise am Anfang
L = [1,["Hello",2.14,-10]]
L1 = copy.deepcopy(L)
L1[1][1] = 6
print(f" L1[1][1]: {L[1][1]}  L2[1][1]: {L1[1][1]}")
````

Wenn man mit verschachtelten Listen arbeitet und dort Kopiervorgänge durchführen möchte, sollte man sich mit copy.deepcopy auf der sicheren Seite bewegen. Kommen wir nun zu den sehr ähnlichen Tupel.
