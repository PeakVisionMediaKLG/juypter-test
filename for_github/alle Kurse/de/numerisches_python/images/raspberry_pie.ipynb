```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

pie_chart = mpimg.imread('pie_chart.png')

plt.axis("off")
plt.imshow(pie_chart)

print(pie_chart[:3]) # pixels are 4 bytes (transparent mode)

pie_chart = pie_chart[:, :, :3]
```

```python
raspberries = mpimg.imread('Raspberries.png')

plt.axis("off")
plt.imshow(raspberries)

print(raspberries[:10])
```

```python
dark_green_area = mpimg.imread("python_dark_green_area.png")

img = np.where(pie_chart > [0.1, 0.1, 0.1], 
                raspberries, 
                dark_green_area)

plt.axis("off")
plt.imshow(img)
```

```python
mpimg.imsave('pie_chart_with_raspberries.png', img)
```
